### Feature Engineering

```python
app_train_updated = app_train_clean.copy(deep=True)
```

```python
# dummy all string cols
dummy_cols = [col for col in app_train_clean.columns if app_train_updated[col].dtype == object]
app_train_updated = pd.get_dummies(app_train_updated, columns=dummy_cols)
```

```python
def inbalance(df, impute=True):
    '''
    Inputs:
        df - dataframe to split by target
        impute - if True, impute (via bootstrap) new values from underbalanced class;
            otherwise, randomly sample from overbalanced class
    Returns:
        dataframe with classes balanced
    '''
    ones = df[ df.TARGET == 1 ]
    len_ones = len(ones)
    zeros = df[ df.TARGET == 0 ]
    len_zeros = len(zeros)

    if len_ones > len_zeros:
        heavy = ones.copy(deep=True)
        light = zeros.copy(deep=True)
    else:
        heavy = zeros.copy(deep=True)
        light = ones.copy(deep=True)
    
    if impute:
        diff = abs(len_ones - len_zeros)
        for _ in range(diff):
            num = random.randint(0, len(light) - 1)
            new_row = light.iloc[num]
            light = light.append(new_row)
    else:
        heavy = heavy.sample(len(light))

    return heavy.append(light)
```

```python
# deal with class inbalance
impute = False

app_train_updated = inbalance(app_train_updated, impute)
```

### Train/Val Split

```python
y = app_train_updated.pop('TARGET')
X_train, X_test, y_train, y_test = train_test_split(app_train_updated, y, test_size=0.2)
```
